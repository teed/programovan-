# 1
Vytvořte třídu Utulek, který bude obsahovat tyto vlastnosti:

*   Název
*   Maximální kapacita
*   SeznamZvířat

A jeho metody budou:

*  getNazev
*  getMaxKapacita
*  getSeznamZvirat
*  setNazev
*  setMaxKapacita
*  pridatZvire
*  odeberZvire
*  ToString







```python
class Utulek:
    def __init__(self,nazev:str,maxCapacity: int, seznamZvirat = None): # pokud nezadam funguje pokud zadám mam seznam
        self.nazev = nazev
        self.maxCapacity = maxCapacity
        if seznamZvirat is None:
            self.seznamZvirat = []
        else:
            self.seznamZvirat = seznamZvirat
    def getNazev(self):
        return self.nazev
    def getMaxCapacity(self):
        return self.maxCapacity
    def getseznamZvirat(self):
        return self.seznamZvirat
    def setNazev(self,nazev):
        self.nazev = nazev
    def setMaxCapacity(self,maxCapacity):
        self.maxCapacity = maxCapacity
    def pridejZvire(self,zvire):
        if len(self.seznamZvirat) >= self.maxCapacity or zvire in self.seznamZvirat:
            print("Přesažena kapacita útulku nebo je již zvíře v útulku")
        else:
            self.seznamZvirat.append(zvire)
    def odeberZvire(self,zvire):
        if zvire in self.seznamZvirat:
            self.seznamZvirat.remove(zvire)
        else:
            print("Zvíře v seznamu tohoto útulku není")
    def seznamToString(self): # metoda, která přepíše seznam do stringu (nevrací class._Pes)
        text = "\n"
        for zvire in self.seznamZvirat:
            text += "{}\n".format(zvire.getJmeno())
        return text
    def __str__(self):
        return f"Útulek {self.nazev} má maximální kapacitu {self.maxCapacity} a obsahuje: {self.seznamToString()}"
```

```python
utulek1 = Utulek("Jmeno",5)
utulek2 = Utulek("Jmeno",6,["Azor", "amalka"])
print(utulek1)
print(utulek2)
```

Vytvořte třídu Zvire, která bude obsahovat tyto vlasnosti
* Jméno
* Hmotnost

a metody budou pro get a set a ToString a metodu zarvi()

```python
class Zvire:
    def __init__(self,jmeno:str,hmotnost:float):
        self.jmeno = jmeno
        self.hmotnost = hmotnost
    def getJmeno(self):
        return self.jmeno
    def setJmeno(self,jmeno):
        self.jmeno = jmeno
    def getHmotnost(self):
        return self.hmotnost
    def setHmotnost(self,hmotnost):
        self.hmotnost = hmotnost
    def __str__(self):
        return f"Zvíře se jmenuje {self.jmeno} a váží {self.hmotnost}"
    def zarvi(self):
        print("waaaaaaaaa")
```
