Vytvořte aplikaci, do které napíšete 5 vstupů.
Mezi vstupy bude minimálně 1 číslo a všechen text bude o minimální délce 5 znaků. Žádný ze vstupů nesmí obsahovat kombinaci čísla a textu.   (viz cvičení na githubu: isdigit/isalpha) PS: žádný ze vstupů si nemusíte pamatovat. Stačí pouze vypsat 5 validních vstupních hodnot)


````python
#TODO
pocet = 0

while pocet < 5:
    vstup = input("Zadejte minimalne 1 cislo nebo slovo o minimalni delce 5: ")

    if vstup.isdigit():
        print("zadali jste:", vstup)
        pocet += 1

    elif vstup.isalpha() and len(vstup) >= 5:
        print("zadali jste:", vstup)
        pocet += 1

    else:
        print("Zadejte jiny vstup. Nesplnili jste zadnou z podminek")
````

Vytvoř program, který požádá uživatele o zadání počtu bodů (0 - 100).

Na základě počtu bodů program přiřadí známku podle těchto pravidel:
* 90 až 100 bodů: "Výborně (1)"
* 75 až 89 bodů: "Chvalitebně (2)"
* 60 až 74 bodů: "Dobře (3)"
* 50 až 59 bodů: "Dostatečně (4)"
* 0 až 49 bodů: "Nedostatečně (5)"

Pokud uživatel zadá číslo mimo rozsah 0-100, program vypíše chybovou hlášku: "Neplatný počet bodů."

````python
#TODO
body = int(input("Zadej počet bodů (0-100): "))

if body < 0 or body > 100:
    print("Neplatny pocet bodu.")
elif body >= 90:
    print("Znamka: Vyborne (1)")
elif body >= 75:
    print("Znamka: Chvalitebne (2)")
elif body >= 60:
    print("Znamka: Dobre (3)")
elif body >= 50:
    print("Znamka: Dostatecne (4)")
else:
    print("Znamka: Nedostatecne (5)")
````

Vytvoř program, který požádá uživatele o zadání jeho celého jména
Program vypíše:
* Počet znaků ve jménu (bez mezer).
* První písmeno křestního jména a první písmeno příjmení (velkým písmem).
* Celé jméno převrácené (např. "Novák Jan" se stane "naJ kávon").

Pokud jméno obsahuje méně než dvě slova (křestní jméno a příjmení), program vypíše varování: "Zadejte své celé jméno (křestní jméno a příjmení)."

````python
#TODO
jmeno = input("Zadej sve cele jmeno: ").strip()

casti = jmeno.split()

if len(casti) < 2:
    print("Zadejte sve cele jméno (krestni jmeno a prijmeni).")
else:
    
    mezer = jmeno.replace(" ", "")
    print("Počet znaku (bez mezer):", len(mezer))

    inicialy = casti[0][0].upper() + " " + casti[1][0].upper()
    print("Inicialy:", inicialy)

    
    obracene = jmeno[::-1]
    print("Prevracené jmeno:", obracene)
````
