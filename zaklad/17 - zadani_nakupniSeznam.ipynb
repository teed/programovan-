```python
# Pokud máme seznam a chceme vypsat jeho obsah pomocí metody __str__,
# musíme vytvořit metodu, která nám seznam převede na string
class Seznam:
    def __init__(self,seznam:list):
        self.seznam = seznam #vše co je v košíku
    def __str__(self):
        #return self.seznam # nevrací string, ale LIST
        return "{}".format(self.seznam) # to už projde
        #ale co když je vstupem seznamu jiná třída

sez = Seznam(["jablko","hrušky"])
print(sez)

class Polozka:
    def __init__(self,pocet:int,cena:float):
        self.cena = cena
        self.pocet = pocet
    def __str__(self):
        return "cena: {} a počet: {}".format(self.cena,self.pocet)

jabka = Polozka(5,3)
pomelo = Polozka(5,10)
sez = Seznam([jabka,pomelo])
print(sez) # print nám vypíše pouze třídu, je třeba dopsat metodu seznamToString()
#[<__main__.Polozka object at 0x785f7118dc00>, <__main__.Polozka object at 0x785f7118dde0>]

class Seznam:
    def __init__(self,seznam:list):
        self.seznam = seznam #vše co je v košíku
    def seznamToString(self): # metoda, která přepíše seznam do stringu
        text = ""
        for polozka in self.seznam:
            text += "cena: {} a počet: {}\n".format(polozka.cena,polozka.pocet)
        return text
    def __str__(self):
        # musím volat metodu seznamToString(),
        # aby mi vypsala jen string všeho v poli
        return "{}".format(self.seznamToString())
        #return "{}".format(self.seznam) # to nefunguje pro vstup Class

sez = Seznam([jabka,pomelo])
print(sez) # print nám vypíše pouze třídu, je třeba dopsat metodu seznamToString()
```

#1
Vytvořte třídu Polozka, která bude mít tyto atributy:

*   Název
*   Počet
*   Cena za kus

Ta bude obsahovat metody pro get, set, ToString.


```python
class Polozka:
    def __init__(self, nazev, pocet, cena):
        self._nazev = nazev
        self._pocet = pocet
        self._cena = cena

   
    def get_nazev(self):
        return self._nazev

    def get_pocet(self):
        return self._pocet

    def get_cena(self):
        return self._cena

    
    def set_nazev(self, nazev):
        self._nazev = nazev

    def set_pocet(self, pocet):
        self._pocet = pocet

    def set_cena(self, cena):
        self._cena = cena

 
    def __str__(self):
        return f"Název: {self._nazev}, Počet: {self._pocet}, Cena za kus: {self._cena} Kč"
```

#2
Vytvořte třídu Seznam, která bude mít tyto atributy:

*   Název seznamu
*   Seznam položek
*   Maximální útrata (kolik mám u sebe peněz)

Ta bude obsahovat metody pro get, set, ToString a metody pro přidání a odebrání položky seznamu a zjištění ceny celého nákupu.


```python
class Seznam:
    def __init__(self, nazev_seznamu, max_utrata):
        self._nazev_seznamu = nazev_seznamu
        self._polozky = []          
        self._max_utrata = max_utrata

   
    def pridej_polozku(self, polozka):
        self._polozky.append(polozka)


    def odeber_polozku(self, polozka):
        if polozka in self._polozky:
            self._polozky.remove(polozka)

  
    def celkova_cena(self):
        suma = 0
        for p in self._polozky:
            suma += p.get_pocet() * p.get_cena()
        return suma

    
    def __str__(self):
        vystup = f"Seznam: {self._nazev_seznamu}\n"
        vystup += f"Maximální útrata: {self._max_utrata} Kč\n"
        vystup += "Položky:\n"
        for p in self._polozky:
            vystup += " - " + str(p) + "\n"
        vystup += f"Celkova cena: {self.celkova_cena()} Kč"
        return vystup

    
```
