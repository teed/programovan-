# 1
Vytvořte třídu Utulek, který bude obsahovat tyto vlastnosti:

*   Název
*   Maximální kapacita
*   SeznamZvířat

A jeho metody budou:

*  getNazev
*  getMaxKapacita
*  getSeznamZvirat
*  setNazev
*  setMaxKapacita
*  pridatZvire
*  odeberZvire
*  ToString







```python
class Utulek:
    def __init__(self, nazev, maxkapacita, zvirata=None):
        self.nazev = nazev
        self.maxkapacita = maxkapacita
        if zvirata is None:
            self.zvirata = []
        else:
            self.zvirata = zvirata

    def pridej_zvire(self, zvire):
        if len(self.zvirata) < self.maxkapacita:
            if zvire in self.zvirata:
                print(f"{zvire} je již v útulku")
            else:
                self.zvirata.append(zvire)
                print(f"{zvire} byl přidán do útulku")
        else:
            print("Útulek je už plný")

    def odeber_zvire(self, zvire):
        if zvire in self.zvirata:
            self.zvirata.remove(zvire)
            print(f"{zvire} bylo odebráno z útulku")
        else:
            print("Zvíře není v útulku")

    def __str__(self):
        return f"Útulek {self.nazev} s kapacitou {self.maxkapacita}"


```

Použijte třídu útulek a nějaký útulek založte. Přidejte do něj větší počet zvířat než je kapacita, a odeberte jedno zvíře

```python
utulek = Utulek("Nadějelll", 3)
print(utulek)

utulek.pridatZvire("Liska")
utulek.pridatZvire("Jezevec")
utulek.pridatZvire("kocka")
print()
print(utulek)
```
