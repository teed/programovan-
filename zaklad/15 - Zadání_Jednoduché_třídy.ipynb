Vytvořte třídu Člověk, která bude mít atributy:


1.   jméno:str
2.   příjmení:str
3.   telefon:int
4.   email:str
5. adresa:str
6. pojišťovna:str

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše o této osobě




````python
class Clovek:
    def __init__(self, jmeno, prijmeni, telefon, email, adresa, pojistovna):
        self.jmeno = jmeno
        self.prijmeni = prijmeni
        self.telefon = telefon
        self.email = email
        self.adresa = adresa
        self.pojistovna = pojistovna

 
    def get_jmeno(self):
        return self.jmeno

    def get_prijmeni(self):
        return self.prijmeni

    def get_telefon(self):
        return self.telefon

    def get_email(self):
        return self.email

    def get_adresa(self):
        return self.adresa

    def get_pojistovna(self):
        return self.pojistovna

    
    def set_jmeno(self, jmeno):
        self.jmeno = jmeno
        print("Jméno změněno.")

    def set_prijmeni(self, prijmeni):
        self.prijmeni = prijmeni
        print("primeni změněno.")
    def set_telefon(self, telefon):
        self.telefon = telefon
        print("telefon  změněno.")
    def set_email(self, email):
        self.email = email
        print("email změněno.")
    def set_adresa(self, adresa):
        self.adresa = adresa
        print(" adresa změněno.")
    def set_pojistovna(self, pojistovna):
        self.pojistovna = pojistovna
        print("pojistovna změněno.")
    
    def __str__(self):
        return f"""
        jmeno :{self.jmeno}
        primeni :{self.prijmeni}
        telefon :{self.telefon}
        email :{self.email}
        adresa:{self.adresa}
       pojistovna :{self.pojistovna}
"""
clovek1 = Clovek("aneta","Fsgsgg","123456670","svsvvvsv@ddsfs.cz","loman","plpl")
clovek2 = Clovek("aůhneta","Fs§lgsůhkgg","1235456670","svsůkůkvvvsv@ddsfs.cz","loůkman","plůhpl")
print(clovek1)
````
