Vytvořte třídu pojištovna, která bude mít atributy:


1.   název:str
2.   zkratkaPojištovny:str (maximálně 3 znaky)
3. sídlo:str

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše o této pojišťovně.




````python
class Pojistovna:
    def __init__(self, nazev, zkratka, sidlo):
        self.nazev = nazev
        self.set_zkratka(zkratka)
        self.sidlo = sidlo


    def get_nazev(self):
        return self.nazev

    def get_zkratka(self):
        return self.zkratka

    def get_sidlo(self):
        return self.sidlo

    def set_nazev(self, nazev):
        self.nazev = nazev

    def set_zkratka(self, zkratka):
        if len(zkratka) > 3:
            self.zkratka = zkratka[:3]
        else:
            self.zkratka = zkratka

    def set_sidlo(self, sidlo):
        self.sidlo = sidlo


    def __str__(self):
        return (
            f"Pojišťovna:\n"
            f"Název: {self.nazev}\n"
            f"Zkratka: {self.zkratka}\n"
            f"Sídlo: {self.sidlo}"
        )



pojistovna = Pojistovna("Kooperativa", "KOOPERATIVA", "Brno")
print(pojistovna)




````

# Počasí

Vytvořte třídu Pocasi, která bude mít atributy:

1.   vítr:float
2.   povetrnostniPodminky:str
3.   teplota:float
4.   vlhkost:float

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše tomto počasí


````python

class Pocasi:
    def __init__(self, vitr:float, poveternostni_podminky, teplota, vlhkost):
        self.vitr = vitr
        self.poveternostni_podminky = poveternostni_podminky
        self.teplota = teplota
        self.vlhkost = vlhkost

 
    def get_vitr(self):
        return self.vitr

    def get_poveternostni_podminky(self):
        return self.poveternostni_podminky

    def get_teplota(self):
        return self.teplota

    def get_vlhkost(self):
        return self.vlhkost

  
    def set_vitr(self, vitr):
        self.vitr = vitr

    def set_poveternostni_podminky(self, podminky):
        self.poveternostni_podminky = podminky

    def set_teplota(self, teplota):
        self.teplota = teplota

    def set_vlhkost(self, vlhkost):
        self.vlhkost = vlhkost


    def __str__(self):
        return (f"""
            Počasí:
            Vítr: {self.vitr} m/s
            Povětrnostní podmínky: {self.poveternostni_podminky}
            Teplota: {self.teplota} °C
            Vlhkost: {self.vlhkost} %"
        """)






````

````python
pocasi = Pocasi(5.2, "Zataženo", 12.5, 65)

print(pocasi)
````
